## 📥 [2A] Простое чтение CSV-файлов в pandas

Сначала просто прочитаем оба CSV-файла в объекты `DataFrame`, без каких‑либо изменений.

После этого мы узнаем, сколько строк загружено в каждый датасет.

```python
# 🐱 Шаг 2A. Чтение CSV-файлов в pandas

import pandas as pd

# **ИЗМЕНЕНО:** читаем ваш файл с небоскрёбами вместо примеров с мультфильмами
df_skyscrapers = pd.read_csv("data/skyscrapers.csv")

# **ИЗМЕНЕНО:** выводим информацию о загруженных данных
print("✅ Загружено строк в df_skyscrapers:", len(df_skyscrapers))
print("✅ Столбцы:", df_skyscrapers.columns.tolist())
```

## 🧹 [2B] Очистка и переименование столбцов

В исходном CSV-файле есть **технические столбцы**, которые полезны для Викиданных, но мешают простому анализу:

- Столбец `skyscraper` с URL (ссылкой на объект Wikidata) — **удаляем**, так как для базового анализа он не нужен.
- Столбцы `skyscraperLabel`, `countryLabel`, `cityLabel`, `architectLabel` содержат читаемые подписи (названия).

В этом шаге мы:
- **удаляем столбец с URL Wikidata** (`skyscraper`);
- переименовываем `skyscraperLabel → skyscraper`, `countryLabel → country`, `cityLabel → city`, `architectLabel → architect`;
- приводим числовые столбцы (`height`, `floors`, `constructionYear`) к типу `int`.

При приведении к числам мы используем:

- `pd.to_numeric(..., errors="coerce")` — преобразует значения в числа, некорректные значения превращает в `NaN`;
- `fillna(0)` — заменяет пропущенные значения (`NaN`) на 0;
- `astype(int)` — переводит столбец к целочисленному типу.

> ⚠️ **Важно:** если в ваших данных есть столбцы вида `*Label`, этот шаг обязателен, чтобы получить аккуратные таблички для анализа. Столбец с URL мы удаляем, так как в данном проекте ссылки на Викиданные не требуются.

```python
# 🧹 Шаг 2B. Очистка и переименование столбцов

# Работаем с датафреймом df_skyscrapers
if "skyscraperLabel" in df_skyscrapers.columns:  # **ИЗМЕНЕНО:** проверка под ваш файл
    # 1) Удаляем технический столбец с URL Wikidata
    # **ИЗМЕНЕНО:** не переименовываем в "URL", а удаляем полностью
    if "skyscraper" in df_skyscrapers.columns:
        df_skyscrapers = df_skyscrapers.drop(columns=["skyscraper"])

    # 2) Переименовываем столбцы с суффиксом Label
    # **ИЗМЕНЕНО:** список переименований под вашу структуру
    df_skyscrapers = df_skyscrapers.rename(columns={
        "skyscraperLabel": "skyscraper",   # ← название небоскрёба
        "countryLabel": "country",         # ← страна
        "cityLabel": "city",               # ← город
        "architectLabel": "architect",     # ← архитектор
    })

    # 3) Приводим числовые столбцы к типу int
    # **ИЗМЕНЕНО:** обрабатываем height, floors, constructionYear
    numeric_cols = ["height", "floors", "constructionYear"]
    for col in numeric_cols:
        if col in df_skyscrapers.columns:
            df_skyscrapers[col] = pd.to_numeric(
                df_skyscrapers[col], errors="coerce"
            ).fillna(0).astype(int)

    print("✅ df_skyscrapers очищен и переименован")
else:
    print("⏭️ df_skyscrapers уже очищен, пропускаем")

print("\n✅ Данные готовы к анализу")
```

## 🔍 [3] Обзор данных: структура и первые строки

Сделаем короткий обзор датафрейма с данными о небоскрёбах:

- посмотрим размер таблицы (`shape`);
- выведем список столбцов;
- посмотрим первые несколько строк;
- дополнительно посчитаем базовую статистику по числовым полям (`height`, `floors`, `constructionYear`).

Для удобства напишем маленькую функцию `show_info(df, name)`, чтобы не повторять один и тот же код при анализе разных таблиц.

```python
def show_info(df, name, n=5):
    """Краткий обзор DataFrame: имя, размер, список столбцов и первые строки."""
    print(f"\n📊 {name}")
    print("Размер:", df.shape)
    print("Столбцы:", ", ".join(df.columns))
    print("\nПервые строки:")
    print(df.head(n))

# 🔍 Шаг 3. Обзор данных

# **ИЗМЕНЕНО:** работаем только с вашим датафреймом df_skyscrapers
show_info(df_skyscrapers, "Небоскрёбы: названия, высота, этажи, локация (df_skyscrapers)")

# **ДОБАВЛЕНО:** базовая статистика по числовым столбцам
# **ИЗМЕНЕНО:** вместо capital_cost анализируем height, floors, constructionYear
print("\n📈 Базовая статистика по числовым полям:")
numeric_cols = ["height", "floors", "constructionYear"]
print(df_skyscrapers[numeric_cols].describe())
```

## ✅ [4] Быстрая проверка и валидация данных

Здесь мы посмотрим:

- сколько **уникальных** небоскрёбов, стран, городов и архитекторов есть в данных;
- **какие страны встречаются чаще всего** (Топ‑5 по числу записей);
- **какие архитекторы спроектировали больше всего зданий** (Топ‑5 по числу строк);
- **диапазон высот и годов постройки**: минимальные/максимальные значения.

Функция `value_counts()`:
- считает, сколько раз каждое значение встречается в столбце;
- сортирует результаты по убыванию.

Метод `.head()` берёт первые N строк, поэтому
`df_skyscrapers["country"].value_counts().head()` даёт **Топ‑5 стран по числу небоскрёбов**.

```python
# ✅ Шаг 4. Быстрая проверка и валидация данных

print("🔍 Быстрая проверка данных")

# **ИЗМЕНЕНО:** работаем только с df_skyscrapers
print("\n📊 Датасет: Небоскрёбы — высота, этажи, локация, архитекторы")

# Уникальные значения по ключевым столбцам
print("Уникальных небоскрёбов:", df_skyscrapers["skyscraper"].nunique())
print("Уникальных стран:", df_skyscrapers["country"].nunique())
print("Уникальных городов:", df_skyscrapers["city"].nunique())
print("Уникальных архитекторов:", df_skyscrapers["architect"].nunique())

# **ИЗМЕНЕНО:** Топ-5 стран и архитекторов вместо жанров
print("\n🌍 Топ-5 стран по числу записей:")
print(df_skyscrapers["country"].value_counts().head())

print("\n👷 Топ-5 архитекторов по числу проектов:")
print(df_skyscrapers["architect"].value_counts().head())

# **ИЗМЕНЕНО:** анализируем constructionYear вместо publicationYear
print("\n📅 Диапазон годов постройки:",
      df_skyscrapers["constructionYear"].min(), "—",
      df_skyscrapers["constructionYear"].max())

# **ДОБАВЛЕНО:** статистика по высотам и этажам
print("\n📏 Статистика по высоте (height):")
print(df_skyscrapers["height"].describe())

print("\n🏢 Статистика по количеству этажей (floors):")
print(df_skyscrapers["floors"].describe())

print("\n📊 Статистика по годам постройки:")
print(df_skyscrapers["constructionYear"].describe())
```
